```python
import pandas as pd
df=pd.read_csv('/content/data/citywise_familywelfare.csv')
df.head()
```

```python
df.tail()
```

```python
df.isna().sum()
```

```python
df.tail()
```

```python
age_mean_value=df['doctorCount'].mean()
df.head()
```

```python
df.tail()
```

```python
df["doctorCount"]=df['doctorCount'].fillna(age_mean_value)
df.head()
```

```python
df.tail()
```

```python
df.drop("wellnessCentreAddress",axis=1,inplace=True)
df.head()
```

```python
df.tail()
```

```python
filtered_age=df[df.doctorCount>10]
filtered_age
```

```python
sorted_place=df.sort_values("cityName",ascending=True,kind='heapsort')
sorted_place.head(10)
```

```python
merged_df=pd.merge(df.head(2),df.tail(2),how='outer',indicator='True')
merged_df
```

```python
group_df=df.groupby('cityName')
for key,item in group_df:
  print(group_df.get_group(key),"\n\n")
```

```python
import pandas as pd
df=pd.read_csv('/content/data/nyc_weather.csv')
df.head()
```

```python
df.isna().sum()
```

```python
cc_mean_value=df['CloudCover'].mean()
df.head()
```

```python
df["CloudCover"]=df['CloudCover'].fillna(cc_mean_value)
df.head()
```

```python
df.drop("Events",axis=1,inplace=True)
df.head()
```

```python
filtered_vm=df[df.WindSpeedMPH>10]
filtered_vm
```

```python
sorted_hm=df.sort_values("Humidity",ascending=True,kind='heapsort')
sorted_hm.head(10)
```

```python
merged_df=pd.merge(df.head(2),df.tail(2),how='outer',indicator='True')
merged_df
```

```python
group_df=df.groupby('Temperature')
for key,item in group_df:
  print(group_df.get_group(key),"\n\n")
```
